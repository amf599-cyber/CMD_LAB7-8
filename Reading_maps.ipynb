The file with the maps contain 15,000 images, and each image has 256x256 pixels.

The files with the parameters has 1,000 entries, and each entry has 6 parameter values.

```python
print(maps.shape)
print(params.shape)
```

Plot the image 5328

```python
plt.imshow(np.log10(maps[5328]),cmap=plt.get_cmap('binary_r'), origin='lower', interpolation='bicubic')
plt.show()
```

The reason why there are 15,000 maps, and only 1,000 combinations of parameters, is because for each simulation (that has some unique values of the cosmological and the astrophysical parameters) we generate 15 maps. Thus, the first 15 maps have the same value of the parameters, the next 15 also share the value of the parameters and so on. Thus, to determine the value of the parameters for a given map one needs to do

```python
map_number = 478
params_map = params[map_number//15]

plt.imshow(np.log10(maps[map_number]),cmap=plt.get_cmap('binary_r'), origin='lower', interpolation='bicubic')
plt.show()
print('Value of the parameters for this map')
print('Omega_m: %.5f'%params_map[0])
print('sigma_8: %.5f'%params_map[1])
print('A_SN1:   %.5f'%params_map[2])
print('A_AGN1:  %.5f'%params_map[3])
print('A_SN2:   %.5f'%params_map[4])
print('A_AGN2:  %.5f'%params_map[5])
```
